# Création de la boucle de nettoyage de données

```python
clean_data=pd.DataFrame(columns=['datetime_debut', 'duree_s', 'conso_W']) # créer un df vide en définissant les colonnes

for i in range(len(raw_data["load_curve"])) :
    data = raw_data["load_curve"][i] #Extraire les infos d'une ligne
    data = data.strip('[]').split('],[') 
    data = [d.split(',') for d in data] #Remettre sous forme de liste de liste
    data = [[datetime.fromtimestamp(int(d[0]) / 1000), int(d[1]), int(d[2])] for d in data] #Convertir le timestamp en Datetime
    data_df= pd.DataFrame(data, columns=['datetime_debut', 'duree_s', 'conso_W']) #Repasser en dataframe
    clean_data=pd.concat([clean_data,data_df], ignore_index=True)# Concatener en DF commun
```

```python
clean_data.info()
```

```python
clean_data.duree_s = clean_data.duree_s.astype(int) # passer la colonne duree_s en integer
clean_data.conso_W = clean_data.conso_W.astype(int)# passer la colonne conso_W en integer
clean_data
```

```python
clean_data.info()
```

# Création de colonnes pour préanalyse

```python
#Tri du dataframe + reset index 
clean_data=clean_data.sort_values(by=['datetime_debut'])
clean_data=clean_data.reset_index(drop=True)
clean_data
```

```python
#Création colonne diff et cumul_diff
clean_data['date']=clean_data['datetime_debut'].dt.date
clean_data['time']=clean_data['datetime_debut'].dt.time
clean_data['day_of_week']=clean_data['datetime_debut'].dt.dayofweek
clean_data['week']=clean_data['datetime_debut'].dt.isocalendar().week
clean_data['year']=clean_data['datetime_debut'].dt.year
clean_data['conso_diff']=clean_data['conso_W']-clean_data['conso_W'].shift(periods=1, fill_value=406)
clean_data['cumul_conso_diff']=clean_data['conso_diff']+clean_data['conso_diff'].shift(periods=1, fill_value=0)
clean_data
```

# Analyse

```python
# Définition des valeurs de seuils à partir des histogrammes
diff_haut = 2300
diff_bas = -2300
conso_haut = 3500
conso_seuil = 2850
conso_bas = 700
```

```python
# Création d'une colonne 1/0 indiquant si la charge est supérieure au seuil défini
conditions = [
    (clean_data['conso_W'] >= conso_seuil),
    (clean_data['conso_W'] < conso_seuil),
    ]

values = [1, 0]

clean_data['charge>seuil'] = np.select(conditions, values, default=np.nan)
```

```python
#Création des colonnes avec décalages de lignes pour faire apparaitre les durées des plages de charge
clean_data['diff_charge_seuil']=clean_data['charge>seuil']-clean_data['charge>seuil'].shift(periods=1, fill_value=0)
clean_data['s_charge_1h']=clean_data['charge>seuil']+clean_data['charge>seuil'].shift(periods=-1, fill_value=0)
clean_data['charge_1h']=clean_data['s_charge_1h']+clean_data['diff_charge_seuil']
clean_data['s_charge_1h30']=clean_data['charge>seuil']+clean_data['charge>seuil'].shift(periods=-1, fill_value=0)+clean_data['charge>seuil'].shift(periods=-2, fill_value=0)
clean_data['charge_1h30']=clean_data['s_charge_1h30']+clean_data['diff_charge_seuil']
clean_data['s_charge_2h']=clean_data['charge>seuil']+clean_data['charge>seuil'].shift(periods=-1, fill_value=0)+clean_data['charge>seuil'].shift(periods=-2, fill_value=0)+clean_data['charge>seuil'].shift(periods=-3, fill_value=0)
clean_data['charge_2h']=clean_data['s_charge_2h']+clean_data['diff_charge_seuil']
```

```python
clean_data
```

```python
# Compte du nombre de période continue de 2h dépassant le seuil de charge
print(clean_data[clean_data['charge_2h']== 5].count())
print(clean_data[clean_data['charge_1h30']== 4].count())
print(clean_data[clean_data['charge_1h']== 3].count())
```

```python
clean_data['is_charging_2h']=np.nan

trigger = 0
for i,v in clean_data['charge_2h'].items():
    if v==5:
        trigger=v
        clean_data['is_charging_2h'][i]=1
    elif v==-1 :
        trigger=v
        clean_data['is_charging_2h'][i]=0
    elif v<5 and v>=0 :
        if trigger==5 and clean_data['charge>seuil'][i]==1:
            clean_data['is_charging_2h'][i]=1
        elif trigger==-1 and clean_data['charge>seuil'][i]==0:
            clean_data['is_charging_2h'][i]=0
        else : 
            clean_data['is_charging_2h'][i]=0
    else : 
        clean_data['is_charging_2h'][i]=0
```

En analysant les périodes d'absence du domicile, on remarque une conso en veille (résiduelle) entre 400 et 500. On prendra 450 pour neutraliser la conso de veille et 3400 comme valeur maxi de conso pour la borne de recharge 

```python
conso_veille=450
conso_max_rech=3400
```

```python
conditions = [
    (clean_data['conso_W'].between(conso_veille,conso_max_rech))  ,
    (clean_data['conso_W'] < conso_veille) ,
    (clean_data['conso_W'] > conso_max_rech)
    ]

values = [clean_data['conso_W']-conso_veille, 0, conso_max_rech]

clean_data['conso_active'] = np.select(conditions, values, default=np.nan)
```

```python
clean_data['conso_rechargeVE']=clean_data['conso_active']*clean_data['is_charging_2h']
clean_data['duree_rechargeVE']=clean_data['duree_s']*clean_data['is_charging_2h']
```

```python
trigger = clean_data['charge_2h']==5
clean_data['n_recharge'] = trigger.cumsum()
clean_data['n_recharge'] = clean_data['n_recharge']*clean_data['is_charging_2h']
```

Ajout d'une colonne HP/HC

```python
debut_hc = datetime.strptime('22:00:00','%H:%M:%S').time()
fin_hc = datetime.strptime('05:00:00','%H:%M:%S').time()
```

```python
conditions = [
    (clean_data['time'] >= debut_hc),
    (clean_data['time'] <= fin_hc),
    (clean_data['time'].between(fin_hc,debut_hc, inclusive='neither')),
    ]

values = ['HC','HC','HP']

clean_data['HP/HC'] = np.select(conditions, values, default=np.nan)
```
